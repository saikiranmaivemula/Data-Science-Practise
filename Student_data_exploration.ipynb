```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df = pd.read_csv(r"/Users/saikiranmaivemula/Desktop/DataScientist/Data Projects/Students/student/student-mat.csv",sep=';')
df.head()
```

```python
df['sex'].value_counts()
```

```python
df.groupby('sex')['G3'].mean()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='studytime', y='G3', data = df)
plt.title('study time vs final grade - G3')
plt.xlabel('study time (1 = <2hrs, 4= 10+ hours)')
plt.ylabel('Final grade (G3)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='absences', y='G3', data = df)
plt.title('Absences vs Final grade (G3)')
plt.xlabel('Number of Absences')
plt.ylabel('Final Grade (G3)')
plt.show()
```

```python

numeric_df = df.select_dtypes(include=['float64', 'int64'])

plt.figure(figsize=(12,10))
sns.heatmap(numeric_df.corr(), annot=True, cmap ='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
